```python
import librosa
import numpy as np
import pickle
import librosa.display
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import sys
sys.path.insert(0, '../')
import irms
```

```python
# Load example data
y, sr = librosa.load('example.wav', sr=44100)
```

```python
# Load model for detection of musical subregions
detector = pickle.load(open('../model/renyi_shannon_model.sav', 'rb'))
```

### Single level IRMS

```python
# Choose your parameters:

# Initial STFT
n_fft = 512
hop_size = 512

# Refinement parameters
k = 5
pct = 50
subregion_size = [800, 800]
```

```python
# Perform the IRMS
irms_ds = irms.irms(y, k, subregion_size, detector, pct, n_fft=n_fft, hop_size=hop_size, normalize=True)
# Generate a 2D visualization
irms_vis = irms_ds.generate_visualization()
```

```python
# Compute the initial spectrogram for comparison
singleres_spec = np.abs(librosa.stft(y, n_fft=n_fft, hop_length=hop_size))
```

```python
plt.figure(figsize=[8,4])

plt.subplot(1,2,1)
plt.title("Initial spectrogram")
librosa.display.specshow(librosa.amplitude_to_db(singleres_spec, ref=np.max), y_axis='log', sr=44100)

plt.subplot(1,2,2)
plt.title("IRMS (50% refined)")
librosa.display.specshow(librosa.amplitude_to_db(irms_vis, ref=np.max), y_axis='log', sr=44100)

plt.tight_layout()
```
